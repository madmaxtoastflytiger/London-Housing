**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great!

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# makign a function to calculate the price ratio of house prices in 2018 vs 1998
def create_price_ratio(df_subset_borough):
    # get only the list of values from the avg price column subset from that year, and convert them into floats
    y1998 = float(df_subset_borough['Average_price'][df_subset_borough['Year']==1998].iloc[0])  # need to write it as y1998 with the tring y in front, as if 1998 then will be an int. Also .iloc[0] to safely get the first element
    y2018 = float(df_subset_borough['Average_price'][df_subset_borough['Year']==2018].iloc[0])
    # then divide the 2 lists to make a list of ratios
    ratio = [y2018/y1998]
    return ratio
```

```python
# creating the requested dfg
dfg = df_analysis.drop(['ID','Month'], axis = 1)
dfg = dfg.groupby(by=['London_Borough', 'Year']).mean()
dfg.head()
```

```python
# reset the index
dfg = dfg.reset_index()
dfg.head()
```

```python
# test if the function works
print(create_price_ratio(dfg[dfg['London_Borough']=='Camden']))
```

```python
# use the new function to make a df of borough and avg price ratio of 2018 / 1998
d_borough_ratio = {}
# make a list containing each unqiue cities once, and for loop through them
for unqiue_borough in dfg['London_Borough'].unique():
    # for each unique borough make a subset of that data
    df_subset_borough = dfg[dfg['London_Borough'] == unqiue_borough]
    # for each unique borough, add a new dictionary set, where key is 'unqiue_borough' and the value is the return from the function
    d_borough_ratio[unqiue_borough] = create_price_ratio(df_subset_borough)

# convert dictionary into a df
df_borough_ratio = pd.DataFrame(d_borough_ratio)
df_borough_ratio.head()
```

```python
# reset the index so we can include the boroughs
df_borough_ratio = df_borough_ratio_T.reset_index()
df_borough_ratio.head()
```

```python
# edit the columns names of the df
df_borough_ratio.rename(columns={'index':'Borough', 0:'avg_price_ratio_2018:1998'}, inplace=True)
df_borough_ratio.head()
```
